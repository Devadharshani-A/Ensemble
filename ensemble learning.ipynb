```python
df.drop(columns=['id','Departure Delay in Minutes', 'Arrival Delay in Minutes'],inplace=True)
```

### x,y split

```python
#split data into inputs and targets
X=df.drop(['satisfaction'],axis=1)
y=df['satisfaction']
```

### train test split

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=30)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### knn

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
#create new a knn model
knn = KNeighborsClassifier()
#create a dictionary of all values we want to test for n_neighbors
params_knn = {'n_neighbors': np.arange(1, 25)}
#use gridsearch to test all values for n_neighbors
knn_gs = GridSearchCV(knn, params_knn, cv=5)
#fit model to training data
knn_gs.fit(X_train, y_train)

#save best model
knn_best = knn_gs.best_estimator_
#check best n_neigbors value
print(knn_gs.best_params_)

print("KNN_best: ",knn_best.score(X_test, y_test)*100," %")
```

###  svm

```python
from sklearn import svm
SVC_rbf = svm.SVC(kernel='rbf').fit(X_train, y_train)
print("Accuracy Radial Basis Kernel:", SVC_rbf.score(X_test, y_test)*100,"%")
```

### Logistic

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

print("Logistic Regression accuracy: ",(log_reg.score(X_test, y_test)*100),"%")
```

### MLP

```python
from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(5,4),
                    random_state=4,
                    verbose=True,
                    learning_rate_init=0.03)
mlp_clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
res = mlp_clf.predict(X_test)
res_acc = accuracy_score(y_test, res)
```

```python
print ('MLP Accuracy : ',res_acc*100,'%')
```

### Ensemble method

```python
from sklearn.ensemble import VotingClassifier
estimators=[('knn', knn_best), ('SVC', SVC_rbf), ('log_reg', log_reg),('MLP',mlp_clf)]
ensemble = VotingClassifier(estimators, voting='hard')
```

```python
ensemble.fit(X_train, y_train)
```

```python
print("Ensemble score: ",ensemble.score(X_test, y_test)*100)
```
